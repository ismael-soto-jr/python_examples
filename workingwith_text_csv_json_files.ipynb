This code reads from a text file (use gettysburg.txt) and prints out:

    The number of words
    Average length of words
    Number of lines (not sentences, just anything divided by a newline)

Includes proper exception handling/catching.

```python
try:
    with open('gettysburg.txt', 'r') as f:
        contents = f.read()
        print(contents)
        #number of words
        num_words = contents.split()
        print('Total number of words :', len(num_words))
        #Avg lenght of wrds
        wrd_count = 0
        letter_count = 0
        for word in contents.split():
            letter_count = letter_count + len(word)
            wrd_count = wrd_count + 1
        print('avg wrd lenght:', round(letter_count/wrd_count, 1))
        
        Counter = 0

        CoList = contents.split("\n")

        for i in CoList:
            if i:
                Counter += 1

        print("This is the number of lines in the file:", Counter)
except PermissionError:
    print('file permission error')
except FileNotFoundError:
    print('does not exist')
f.close()
```

This section reads from the file words.txt, then creates a new file containing only 5 letter words. 
Includes proper exception handling.
Also, prints the number of words in the original file and the number of words in the new file.
(N.B. Watch out for newlines.)

```python
try:
    with open('words.txt', 'r') as f:
        wrds_text = f.read()
        number_of_words = 0
        lines = wrds_text.split()
        number_of_words += len(lines)
        print('number of words in the original file:', number_of_words)
except PermissionError:
    print('file permission error')
except FileNotFoundError:
    print('does not exist')
f.close()
```

This section reads a CSV file (sf_buildings.csv) and then:

    Removes the 'use' column
    Converts the height from feet to meters
    Writes the new data to a new CSV file

```python
import csv
```

```python
try:  
    with open('sf_buildings.csv', 'r') as file:
        rdr = list(csv.reader(file))
        hdr = rdr[0]
        for row in rdr:
            print(row)
except PermissionError:
    print('file permission error')
except FileNotFoundError:
    print('does not exist')
file.close()
```

```python
 try:   
    with open('sf_buildings.csv', 'r') as csv_file:
        csv_reader = csv.DictReader(csv_file)

        with open('new_sf_buildings.csv', 'w') as new_file:
            fieldnames = ['name', 'height', 'floors', 'year']

            csv_writer = csv.DictWriter(new_file, fieldnames=fieldnames, delimiter='\t')

            csv_writer.writeheader()

            for line in csv_reader:
                del line['use']

                csv_writer.writerow(line)
except PermissionError:
    print('file permission error')
except FileNotFoundError:
    print('does not exist')
csv_file.close()
new_file.close()

    
```

```python
with open('new_sf_buildings.csv', 'r') as file:
    reader2 = csv.DictReader(file, delimiter='\t')
    for row in reader2:
        print(row)
```

# JSON

Here we read from a JSON file (us_states_and_cities.json) and then:
    
    Print the name of the state that has the most cities and the number of cities it has


```python
import json

try:
    with open('us_states_and_cities.json', 'r') as json_file:
        us_file = json.load(json_file)
    print(us_file)
    print(type(us_file))
    print(json.dumps(us_file, indent=4, sort_keys=True))
except PermissionError:
    print('file permission error')
except FileNotFoundError:
    print('does not exist')
json_file.close()
```

```python
for (k, v) in us_file.items():
    print("Key: " + k)
    print("Value: " + str(len(v)))
```

```python
max_cities = 0
for (k, v) in us_file.items():
    num_cities = len(v)
     
    if num_cities >= max_cities:
        max_cities = len(v)
        state = str(k)
print('State with most cities:', state, 'Number of cities:',max_cities)
    
    
```
